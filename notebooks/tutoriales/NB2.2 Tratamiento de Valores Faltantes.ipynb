## Simple Imputer

```python
from sklearn.impute import SimpleImputer
```

```python
dataframe = diabetes_dataframe.copy()
```

```python
dataframe.head()
```

```python
values = dataframe.values
values
```

```python
values[2]
```

```python
# 1. Creación de los objetos
imputer = SimpleImputer(missing_values=nan, strategy='mean')

# 2. Ajustar a los datos
imputer.fit(values)

# 3. Transformación
transformed_values = imputer.transform(values)

#transformed_values = imputer.fit_transform(values)
```

```python
transformed_values
```

```python
dataframe_transformed = pd.DataFrame(transformed_values)
dataframe_transformed.head()
```

## KNN Imputer

```python
from sklearn.impute import KNNImputer
```

```python
dataframe = diabetes_dataframe.copy()
```

```python
values = dataframe.values
```

```python
type(dataframe)
```

```python
type(values)
```

```python
# 1. Crear el objeto
imputer = KNNImputer(n_neighbors=3, weights='distance', metric='nan_euclidean')

# 2. Ajustar a los datos
imputer.fit(values)

# 3. Transformación
transformed_values = imputer.transform(values)

#transformed_values = imputer.fit_transform(values)
```

```python
values[2]
```

```python
transformed_values[2]
```

```python
dataframe_transformed = pd.DataFrame(transformed_values)
dataframe_transformed.head()
```
